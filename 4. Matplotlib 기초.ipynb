```python
# matplotlib 라이브러리를 불러옵니다. 
%matplotlib inline
import matplotlib.pyplot as plt
# 판다스 라이브러리 불러옵니다.
import pandas as pd 

names = ['Bob','Jessica','Mary','John','Mel']
births = [968, 155, 77, 578, 973]
custom = [1, 5, 25, 13, 23232]

BabyDataSet = list(zip(names, births))
df = pd.DataFrame(data = BabyDataSet, columns=['Names','Births'])

y = df['Births']
x = df['Names']

# 막대 그래프를 출력합니다.
plt.bar(x,y) #막대 그래프 개체 생성
plt.xlabel('Names') #그래프 x축 제목
plt.ylabel('Births') # 그래프 y 축 제목
plt.title('Bar plot') # 그래프 제목
plt.show() # 그래프 출력
```

```python
# 넘파이 라이브러리를 불러옵니다.
import numpy as np

# 랜덤추출시드
np.random.seed(19930128)

# 산점도 데이터를 생성합니다.
# 0 ~ 100까지 5의 범위로 x축을 지정
x = np.arange(0.0, 100.0, 5.0)
y = (x * 1.5) + np.random.rand(20) * 50

# 산점도 데이터를 출력합니다.
plt.scatter(x, y, c = "b", alpha=0.5, label="scatter point")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(loc='upper left')
plt.title('Scatter plot')
plt.show()
```
